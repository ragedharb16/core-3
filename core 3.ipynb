```python
df1[["Hero", "Publisher"]] = df1['Hero|Publisher'].str.split("|", expand =True)
df1.head()
```

```python
df1 = df1.drop(columns = ["Hero|Publisher"])
df1.head(0)
```

```python
#Replacing the single quotation with the double quotation mark
df1['Measurements'] = df1['Measurements'].str.replace("'", '"')
df1
```

```python
probe2= df1.loc[0,'Measurements']
probe2
```

```python
#Applying .json.loads to the entire column
df1['Measurements'] = df1["Measurements"].apply(json.loads)
```

```python
#Converting into a series
df1[["Height(cm)", "Weight(kg)"]] = df1["Measurements"].apply(pd.Series)
df1.drop(columns='Measurements', inplace=True)
```

```python
df1
```

```python
replace_list = [" ", "cm"]
for char in replace_list:
    
    df1["Height(cm)"] = df1["Height(cm)"].str.replace(char, "", regex= False)
```

```python
#Converting column into a numeric format (int)
df1["Height(cm)"] =pd.to_numeric(df1["Height(cm)"]).astype(int)
df1["Height(cm)"].dtype
```

```python
#Getting rid of all white spaces and measurement units using a loop
replace_list = [" ", "kg"]
for char in replace_list:
    
    df1["Weight(kg)"] = df1["Weight(kg)"].str.replace(char, "", regex= False)
    
```

```python
df1["Weight(kg)"] =pd.to_numeric(df1["Weight(kg)"]).astype(int)
df1["Weight(kg)"].dtype
```

```python
df2['Powers_split'] = df2['Powers'].str.split(',')
df2.head()
```

```python
superpowers = df2["Powers"].str.split(",")
superpowers
```

```python
superpowers.explode("Powers")
superpowers.explode("Powers").value_counts()
```

```python
new_column = superpowers.explode("Powers").dropna().unique()
new_column
```

```python
#One Hot Encoding the newly made colums of super hero powers
for col in new_column:
    df2[col] = df2['Powers'].str.contains(col)
df2.head(3)
```

```python
df2= df2.drop(columns = ["Powers","Powers_split"])
```

```python
df2 = df2.rename(columns = {"hero_names": "Hero"})
df2.head()
```

```python
heros_df =pd.merge(df1, df2, how="outer", on="Hero")
```
